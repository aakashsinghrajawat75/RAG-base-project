### read_pdf
* def read_pdf(bytes_blob: bytes) -> str: Defines a function that takes PDF data in bytes format and returns extracted text as a string.
* PdfReader(io.BytesIO(bytes_blob)): Converts the PDF bytes into a file-like object and opens the PDF for reading.
* text = []: Creates an empty list to store text from each page.
* for page in reader.pages: Loops through all pages in the PDF.
* try: Starts error handling to prevent the program from stopping if text extraction fails on a page.
* page.extract_text(): Extracts text from the current page.
* text.append(page.extract_text() or ""): Adds the extracted text to the list. If no text is found, an empty string is added.
* except: Catches any errors that occur during text extraction.
* text.append(""): Adds an empty string when extraction fails, allowing the process to continue.
* "\n".join(text): Combines text from all pages into a single string, separated by new lines.
* return "\n".join(text): Returns the complete extracted text from the PDF.

```python
def read_pdf(bytes_blob:bytes) -> str:
  reader = PdfReader(io.BytesIO(bytes_blob))
  text = []
  for page in reader.pages:
    try:
      text.append(page.extract_text() or "")
    except:
      text.append("")
  return "\n".join(text)
```

### read_docx
* bytes_blob: Contains the DOCX file data in bytes format.
* io.BytesIO(bytes_blob): Converts DOCX bytes into a file-like object.
* DocxDocument(fh): Opens and reads the DOCX document.
* doc.paragraphs: Accesses all paragraphs in the document.
* loop: Extracts text from each paragraph and stores it in a list.

```python
def read_docx(bytes_blob:bytes) -> str:
  fh = io.BytesIO(bytes_blob)
  doc = DocxDocument(fh)
  return "\n".join([p.text for p in doc.paragraphs])
```

### read_txt() Function
* bytes_blob: Contains the TXT file data in bytes format.
* io.BytesIO(bytes_blob): Converts the bytes into a file-like object.
* Uses the decode('utf-8') method to convert bytes into readable text.
* UTF-8 encoding ensures proper conversion of characters and symbols.
* Returns the decoded text as a string.

```python
def read_txt(bytes_blob:bytes) -> str:
  return io.BytesIO(bytes_blob).read().decode('utf-8',errors='ignore')
```

### load_text_by_ext
* fname: Stores the file name andblob: Contains the file data in bytes format
* **os.path.splitext(fname)**: Splits the file name into name and extension
* **[1]**: Retrieves only the file extension & **.lower()**: Converts the extension to lowercase for consistent checking
* if ext == ".txt": Checks if the file is a TXT file
    * read_txt(blob): Extracts text from the TXT file
* elif ext == ".pdf": Checks if the file is a PDF file
    * read_pdf(blob): Extracts text from the PDF file
* elif ext == ".docx": Checks if the file is a DOCX file
    * read_docx(blob): Extracts text from the DOCX file
* else: Handles unsupported file formats
    * ValueError: Raises an error if the file extension is not supported
* return: Returns the extracted text from the corresponding file type

```python
def load_text_by_ext(fname:str, blob:bytes) -> str:
  ext = os.path.splitext(fname)[1].lower()
  if ext == ".txt":
    return read_txt(blob)
  elif ext == ".pdf":
    return read_pdf(blob)
  elif ext == ".docx":
    return read_docx(blob)
  else:
    raise ValueError(f"Unsupported extension: {ext}")
```

### _clean_text
* s: Takes a text string as input
* re.sub(r'\s', ' ', s): Replaces all whitespace characters (spaces, tabs, newlines) with a single space
* .strip(): Removes extra spaces from the beginning and end of the text
* re.sub(r'^\W+|\W+$', '', s): Removes non-word characters (punctuation, symbols, etc.) from the start and end of the text



---



> below 3 function pdf,txt,docx all call this to remove extra space



```python
def _clean_text(s:str) ->str:
  s = re.sub(r'\s',' ',s).strip()
  s= re.sub(r'^\W+|\W+$','',s).strip()
  return s
```

### extract_title_from_pdf_layout


> In this function we are going to get info about document inside text like len, bold, size, case
> page structure page > block > line > spans

**here is some point about that:**
* page_count if their 0 page or empty pdf return none
* spans(list): store info about text like len, bold, size, case
* page_height: store page length
* other info i comment out

> At the last add if any case spans not available then return num


```python
'''
Defines a function to extract the title from a PDF.
pdf_bytes: PDF file content in bytes.
consider_pages=3: Checks only the first 3 pages.
top_band=0.35: Focuses on the top 35% of each pag
'''
def extract_title_from_pdf_layout(pdf_bytes: bytes, consider_pages=3, top_band=0.35):
    doc = fitz.open(stream=pdf_bytes, filetype="pdf")
    if doc.page_count == 0: return None

    spans, page_heights = [], {}
    for pno in range(min(consider_pages, doc.page_count)):
        pg = doc.load_page(pno)
        page_heights[pno] = pg.rect.height
        data = pg.get_text("dict") # in a document have a page >
        for b in data.get("blocks", []):      #   it is file
            for l in b.get("lines", []):      #     structure
                for s in l.get("spans", []):  #   in this pattern avilable text
                    txt = _clean_text(s.get("text","")) # get txt if not it return empty
                    if not txt or len(txt) < 3: continue
                    size = float(s.get("size", 0))     # store text size
                    font = (s.get("font") or "").lower() # font case store
                    is_bold = ("bold" in font) or ("black" in font) # store T/F if bold or not
                    y0 = s.get("bbox", [0,0,0,0])[1]  # store axis if not return [0,0,0,0] for 1
                    spans.append({"page": pno, "text": txt, "size": size, "bold": is_bold, "y0": y0}) # at the end store all info in spans

    if not spans: return None
```

## Title extract form document(.docx)

> in this function we are going to extract title from the docx(document) and we are doing step by step

#### extract_title_from_docx
1. d in convert byte into the document and make it readable
2. we are going to create func that give title from the document
      #### paragraph_scoure
      1. previous func. that remove unnessary space
      2. add a condition that ignore len of text/paragraph
      3. style_name give paragraph style in lower case
      4. run loop on pargraph(p) and inside loop run in font
          * **max_size_pt** store largest font size
          * **any_bold** store if text bold
      5. **style_peiorty** based on basic concept(title > heading 1>heading>)
      6. return text,style_peiorty, size, bold
3. candidates : create list for store tile candidates
4. loop througt all pargraph doc and add into list
5. add a condition if candidate not present by any case return none
6. at second last step sort all candidate
7. return hight probably title


```python
def extract_title_from_docx(docx_bytes:bytes):
  d = DocxDocument(io.BytesIO(docx_bytes))
  def paragraph_score(p):
    text = _clean_text(p.text)
    if not text or len(text) < 3: return None
    style_name = (p.style.name if p.style else "").lower()
    max_size_pt,any_bold = 0.0, False
    for run in p.run:
      if run.font is not None:
        if run.font.size:
          try:
            max_size_pt = max(max_size_pt,float(run.font.size.pt))
          except:
            pass
        if run.font.bold: any_bold = True
    style_priorty = 0
    if "title" in style_name: style_priorty = 3
    elif "heading 1" in style_name: style_priorty = 2
    elif "heading" in style_name: style_priorty = 1
    return {"text":text,"style_priorty":style_priorty, "size" : max_size_pt, "bold": any_bold}


  candidates = []
  for p in d.paragraphs:
    sc = paragraph_score(p)
    if sc: candidates.append(sc)

  if not candidates: return None
  candidates.sort(key=lambda c: (c["style_priorty"], -c["size"], -c["bold"], c["text"]))
  return candidates[0]["text"]
```

## Step 15: Create a sentence-based chunking function with overlap (prepare text for search and Q&A)


**chunk_text** : parameter that in function
* text in string format
* target_chars(int) chunk size(as we need or pc)
* overlap_chars(int): it store connection
* o/p must be list and string format

* **sent_tokenize**: it function in nltk and split text into indiviual sentence
* **chunks** list for store chunks
* buf: it is for chunks buffer

#### Chunk Creation
* loop for all sentence
* **conditions**
    1. if buf is empty or first time loop run
        * store the sentence in buffer
    2. check condition
        * add/append in buf
    3. else lenth croess target size
        * save the current chunk
        * Extracts the last overlap tail
        * ?

* add condition and remove unnessary space and add chunks
* **return** remove empty chunk


```python
def chunk_text(text: str, target_chars: int = 1200, overlap_chars: int = 150) -> List[str]:
    sents = sent_tokenize(text)
    chunks = []
    buf = ""
    for s in sents:
        if not buf:
            buf = s
        elif len(buf) + 1 + len(s) <= target_chars:
            buf += " " + s
        else:
            chunks.append(buf.strip())
            # overlap tail
            tail = buf[-overlap_chars:]
            buf = (tail + " " + s).strip()
    if buf:
        chunks.append(buf.strip())
    # filter tiny empties
    return [c for c in chunks if len(c.strip()) > 0]
```

## Step 17: Create embeddings for all chunks and build a FAISS search index


**Compute embeddings**
* for i in range(0, len(corpus_texts), BATCH):: Processes chunks in batches.
* corpus_texts[i:i+BATCH]: Selects a batch of chunk texts.
* embedder.encode(...): Converts text chunks into vector embeddings.
* normalize_embeddings=True: Normalizes vectors for cosine similarity search.
* show_progress_bar=False: Disables the progress bar.

**Convert to NumPy Array**
* np.vstack(embs): Combines all embeddings into a single matrix.
* .astype('float32'): Converts embeddings to 32-bit floating-point format required by FAISS.

```python
# load the mode
EMBED_MODEL_NAME = "sentence-transformers/all-MiniLM-L6-v2"
# SentenceTransformer is nitl function (i know)
embedder = SentenceTransformer(EMBED_MODEL_NAME, device=DEVICE)

# Flatten chunks + keep metadata
corpus_texts, corpus_meta = [], []  # meta: (doc_id, doc_name, chunk_id)
for doc_id, d in enumerate(all_docs_chunks):   # Iterates all document through doc_id
    for chunk_id, c in enumerate(d["chunks"]): # Iterates throught chunks of the document
        corpus_texts.append(c) # store/ add chunks
        corpus_meta.append((doc_id, d["name"], chunk_id)) # store additional info

# Compute embeddings
BATCH = 64 # it may be different
embs = [] # empty list store embedding
for i in range(0, len(corpus_texts), BATCH):  # it run like range(0,10,3)
    embs.extend(embedder.encode(corpus_texts[i:i+BATCH], normalize_embeddings=True, show_progress_bar=False))

# Convert to NumPy Array
embs = np.vstack(embs).astype('float32')

# FAISS index (cosine ‚Üí use IndexFlatIP with normalized vectors)
index = faiss.IndexFlatIP(embs.shape[1]) # Creates a FAISS index using Inner Product similarity and set dim
index.add(embs) # store all embedding

# Returns the total number of vectors stored in the index.
print("Index built with", index.ntotal, "vectors.")
```

```python
def summarize_long_text(text: str, chunk_chars: int = 1200, overlap: int = 150,
                        map_max_len=256, map_min_len=60) -> str:

    # MAP: summarize each chunk with BART
    parts = chunk_text(text, target_chars=chunk_chars, overlap_chars=overlap)
    partial_summaries = []

    for p in parts if parts else [text]:
        p_in = safe_truncate(p, 3500) # call previous cell
        out = summarizer(p_in, max_new_tokens=map_max_len, min_length=map_min_len, do_sample=False)[0]["generated_text"]
        partial_summaries.append(out.strip())

    # REDUCE: join partials and ask FLAN for crisp bullet points
    joined = safe_truncate(" ".join(partial_summaries), 4500)

    reduce_prompt = f"""Summarize the document below into 7‚Äì10 concise bullet points.
Each bullet must be specific and factual, and should include named entities (people, orgs, places), dates/years,
numbers/tables findings if present, and concrete outcomes or claims. Avoid generic sentences.

Document:
\"\"\"{joined}\"\"\"

Return only bullet points, each starting with '-'."""

    # temperature controls creativity, `temperature = 0.2`is low temperature so that answers are more factual, focused and less random.
    final = generator(reduce_prompt, max_new_tokens=512, temperature=0.2)[0]["generated_text"]

    # Light cleanup: ensure bullets
    lines = [ln.strip() for ln in final.splitlines() if ln.strip()]
    bullets = []
    for ln in lines:
        if not ln.startswith("-"):
            ln = "- " + ln.lstrip("‚Ä¢*- ")
        bullets.append(ln)

    # Keep 7‚Äì10 bullets if possible
    if len(bullets) < 5:
        # fallback to original joined summary split
        bullets = ["- " + s for s in partial_summaries[:8]]
    return "\n".join(bullets[:10]).strip()
```

## Step 22: Build a search function that finds the most relevant chunks using embeddings + FAISS


```python
def search(query: str, top_k: int = 6) -> List[Tuple[float, str, Tuple[int,str,int]]]:
    # This converts the query text into an embedding vector (numbers representing meaning).
    qe = embedder.encode([query], normalize_embeddings=True)

    # This searches the FAISS index to find the closest vectors to the query vector.
          # D: distances/similarity scores for each match
          # I: indices (positions) of the matching vectors in the index
    D, I = index.search(qe.astype('float32'), top_k)

    hits = []
    for score, idx in zip(D[0], I[0]):
        hits.append((float(score), corpus_texts[idx], corpus_meta[idx]))  # (score, chunk_text, (doc_id, doc_name, chunk_id))
    return hits

```

## Step 24: Create a ‚Äúgeneric question detector‚Äù using regex (to improve FAQ quality)




> creates a list of regular expression patterns for identifying generic or low-quality questions



**GENERIC_RE**
* compiles the regular expression for faster searching
* combine all pattern in single expression
* re.I: Makes pattern matching case-insensitive (same as re.IGNORECASE)

```python
# Improved FAQ Generation v2 (varied questions, grounded answers, no placeholders) ===
# helpers to detect/clean generic questions ---
GENERIC_PATTERNS = [
    r"\bmain point\b",                            # ignore "What is the main point of the document?"
    r"\bwhy .* important\b",
    r"\bintended audience\b",
    r"\bscope\b",
    r"\bconclusion\b",                            # What is the conclusion?
    r"\boverview\b",
    r"\bsum(mary|marize)\b",
    r"\badditional detail\??\b",
    r"\bwhat is .*document\b",                   # Why is this document important?
    r"\bwhat does the document\b",
]
GENERIC_RE = re.compile("|".join(GENERIC_PATTERNS), re.I)  # re.IGNORECASE

```

## Step 25: Create a ‚Äúquestion quality checker‚Äù to keep only specific FAQ questions


**looks_specific**
> parameter in string and this function return boolean


* Removes a trailing question mark if present. (rstrip: right side)
* conditions
    1. split the question into word by word and check length
    2. search pattern
    3. Check Vague Openings
        * re.match: check whether the question starts with vague phrases.
        * (?i): Makes the pattern case-insensitive.
        * q[:20]: Examines only the first 20 characters of the question.
* return True: Returns True if the question passes all checks.

```python
def looks_specific(q: str) -> bool:
    q = q.strip().rstrip("?")
    if len(q.split()) < 6:  # avoid very short
        return False
    if GENERIC_RE.search(q):
        return False
    # discourage ultra-vague starts
    if re.match(r"(?i)what is|what does|why is|who is$", q[:20]):
        return False
    return True
```

## Step 26: Remove duplicate items while keeping the original order (clean duplicates smartly)


**dedup_preserve_order**
> items: take a list of items

* seen = set(): to remove dublicate
* out empty list
* loop at items
    * remove splace
    * convert into lower case
    * re.sub(r"\s+", " ", ..) : remove multiple space into one space
    * **condition** if same thing/value then skip
    * add into set and add orignal clean item
* return in orig. order

```python
def dedup_preserve_order(items):
    seen = set(); out = []
    for x in items:
        key = re.sub(r"\s+", " ", x.strip().lower())
        if key in seen:
            continue
        seen.add(key); out.append(x.strip())
    return out
```

## Step 27: Generate high-quality, non-generic FAQ questions from a document summary


**here is some point**
* prompt used to generate FAQ-style questions.

**Function define**
* it takes the document summary as input & max number of return question
* base : store 19 step (limit the len summary)
* prompt : it store prompt as i/p

    **generator** (it store gererated ques)
    * limit the len of o/p
    * increases randemness
    * it uses nucleus sampling for diverse outputs
    * enable random sampling

* qs remove extra space and split line
* remove duplicate question
* keep only document specfic ques
* random ques order to avoid the sequence every runn
* return faq


```python
# --- Question generation from summary (sampling ON for variety) ---
FAQ_QS_PROMPT = """Write 12 distinct, specific FAQ questions that a reader of the following document would likely ask.
Questions must reference concrete people/organizations, locations, years/dates, metrics, methods, or outcomes from the text.
Avoid generic questions like "What is the main point?" or "Why is this important?".

Return ONLY the questions, one per line, no numbering or extra text.

Document:
\"\"\"{doc}\"\"\""""

def generate_varied_questions_from_summary(summary_text: str, k: int = 12) -> list:
    base = safe_truncate(summary_text, 4500)
    prompt = FAQ_QS_PROMPT.format(doc=base)

    out = generator(
        prompt,
        max_new_tokens=256,
        temperature=0.9,       # sampling for variety
        top_p=0.9,
        do_sample=True
    )[0]["generated_text"]


    qs = [q.strip().rstrip("?") + "?" for q in out.splitlines() if q.strip()]

    qs = dedup_preserve_order(qs)
    # filter generic
    qs = [q for q in qs if looks_specific(q)]
    # keep top ~10 and randomize a bit to avoid same 5 every run
    random.shuffle(qs)
    return qs[:max(k, 5)]
```

## Step 28: Answer a question using only one selected document (robust retrieval + fallback)


```python
# --- Answering with robust context (doc retrieval + summary fallback) ---
def answer_within_doc_robust(question: str, doc_name: str, summary_text: str) -> str:
    # primary: restrict hits to this doc
    hits = search(question, top_k=12)
    hits = [h for h in hits if h[2][1] == doc_name]
    ctx_blocks, seen = [], set()
    for sc, txt, meta in sorted(hits, key=lambda x: -x[0]):
        if len(" ".join(ctx_blocks)) > 4200: break
        sig = hash(txt)
        if sig in seen:
            continue
        seen.add(sig); ctx_blocks.append(txt)

    # fallback: if retrieval is weak, include summary
    context = "\n\n".join(ctx_blocks)
    if len(context) < 400 and summary_text:
        # prepend summary to ensure we have something factual
        context = (summary_text.strip() + "\n\n" + context).strip()

    if not context:
        # absolute fallback to summary only (better than ‚ÄúNot specified‚Äù)
        context = summary_text if summary_text else "No context."

    prompt = ANSWER_PROMPT_TMPL.format(context=context, question=question)
    ans = generator(
        prompt,
        max_new_tokens=220,
        temperature=0.2,   # deterministic-ish answers
        do_sample=False
    )[0]["generated_text"].strip()

    # last resort: if model still returns something empty or evasive, try a 2nd pass using only summary
    if (not ans) or len(ans.split()) < 3 or "I don't know" in ans:
        if summary_text:
            prompt2 = ANSWER_PROMPT_TMPL.format(context=summary_text, question=question)
            ans2 = generator(prompt2, max_new_tokens=200, temperature=0.2, do_sample=False)[0]["generated_text"].strip()
            if ans2 and "I don't know" not in ans2:
                ans = ans2

    # keep it tight
    return ans
```

## Step 29: Create an FAQ generator for one document (make 5 good questions and answer them)


```python
# --- Build FAQs using title for display, filename for retrieval mapping ---
def generate_faqs_for_document_v2(doc_name: str, title: str, text: str, summary: str) -> list:
    # ensure we have a seed summary; if not, synthesize a short one
    seed = summary if (summary and len(summary) > 60) else summarize_long_text(text)

    # 1) make 12 candidates from summary
    candidates = generate_varied_questions_from_summary(seed, k=12)

    # 2) if we somehow have <5, synthesize from top bullets
    if len(candidates) < 5:
        bullets = [ln[2:].strip() for ln in seed.splitlines() if ln.strip().startswith("- ")]
        for b in bullets:
            if len(candidates) >= 5: break
            frag = " ".join(b.split()[:12])
            candidates.append(f"What specific findings does the document report about {frag}?")
        candidates = dedup_preserve_order(candidates)

    # 3) pick 5 best-looking questions (stable order but de-generic)
    final_qs = candidates[:5]

    # 4) answer each question robustly
    faqs = []
    for q in final_qs:
        a = answer_within_doc_robust(q, doc_name, summary)
        faqs.append((q, a))
    return faqs
```

## Step 31: Create the main Q&A function (retrieve chunks, build context, generate answer, show sources)




> prompt template used to generate answers from retrieved document context



```python
ANSWER_PROMPT_TMPL = """You are a helpful assistant. Answer the question strictly using the provided context.
If the answer is not in the context, say "I don't know from the provided documents."

Context:
{context}

Question: {question}

Answer:"""
```

**answer_question function**
* hits : it store faiss index relevent chunks
* loop
    * itertes through each result
    * sort decending result simpilar
    * sc: Similarity score
    * txt: Retrieved text chunk
    * meta: Document metadata

* condition for max char limit
* context : it stores the combined context
* prompt: stores the final prompt
* **Generate Answer** (out)
    > Uses the language model to generate an answer
    1. limits the answer length
    2. temperature makes the model highly focused, predictable, and deterministic
    3. gen. ques
    4. remove space

* information collect
* return result
    1. "answer": out: Stores the generated answer
    2. "sources": sources: Stores metadata for the retrieved chunks
    3. "retrieved_snippets": Stores the first 3 retrieved text snippets

```python
def answer_question(question: str, top_k: int = 6, max_ctx_chars: int = 4200) -> Dict:
    hits = search(question, top_k=top_k)

    # merge contexts, sorted by score descending
    ctx_blocks = []
    seen = set()

    for sc, txt, meta in sorted(hits, key=lambda x: -x[0]):
        if len(" ".join(ctx_blocks)) > max_ctx_chars: break
        # avoid exact dupes
        sig = hash(txt)   # create unique hash the text chunks
        if sig in seen:   # if chunk already have
            continue
        seen.add(sig)
        ctx_blocks.append(txt) # add unique chunk
    context = "\n\n".join(ctx_blocks)
    prompt = ANSWER_PROMPT_TMPL.format(context=context, question=question)
    out = generator(prompt, max_new_tokens=256, temperature=0.2)[0]["generated_text"].strip()
    # include source pointers
    sources = [{"doc_name": m[1], "chunk_id": m[2], "score": float(s)} for (s,_,m) in hits[:min(top_k,5)]]
    return {"answer": out, "sources": sources, "retrieved_snippets": [t for _,t,_ in hits[:3]]}
```

## Step 33: Create the Q&A controller for the Gradio UI


```python
# Build mapping: title -> filename, and "All"
TITLE_TO_NAME = {name_to_title[d["name"]]: d["name"] for d in docs_text}
DOC_TITLES = ["All"] + list(TITLE_TO_NAME.keys())

def qa_interface(question, title_scope):
    # map chosen title to original filename
    if title_scope != "All":
        doc_scope = TITLE_TO_NAME[title_scope]
        # Restrict to that doc
        hits = search(question, top_k=12)
        hits = [h for h in hits if h[2][1] == doc_scope]
        ctx = []
        seen = set()
        for sc, txt, meta in sorted(hits, key=lambda x: -x[0]):
            if len(" ".join(ctx)) > 4200: break
            sig = hash(txt)
            if sig in seen: continue
            seen.add(sig); ctx.append(txt)
        context = "\n\n".join(ctx) if ctx else "No context available."
        prompt = ANSWER_PROMPT_TMPL.format(context=context, question=question)
        out = generator(prompt, max_new_tokens=256, temperature=0.2)[0]["generated_text"].strip()
        sources = [{"title": name_to_title[m[1]], "chunk_id": m[2], "score": float(s)} for (s,_,m) in hits[:5]]
        return out, json.dumps(sources, indent=2)
    else:
        res = answer_question(question, top_k=8)
        # decorate sources with titles
        for s in res["sources"]:
            s["title"] = name_to_title.get(s["doc_name"], s["doc_name"])
            del s["doc_name"]
        return res["answer"], json.dumps(res["sources"], indent=2)
```
